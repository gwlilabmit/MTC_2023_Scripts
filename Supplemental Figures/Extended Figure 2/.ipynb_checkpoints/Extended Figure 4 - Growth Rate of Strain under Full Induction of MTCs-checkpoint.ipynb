```python
import pandas as pd
import os
from bokeh.io import push_notebook, show, output_notebook, export_svg, curdoc
from bokeh.plotting import figure
from bokeh.models import LinearColorMapper, ColorBar
import datetime
from scipy.stats import gaussian_kde
import numpy as np
from sklearn.metrics import r2_score
from bokeh.themes import Theme
curdoc().theme = Theme(filename="../../figure_theme.yaml")

output_notebook()
```

```python
data_location = "C:\\Users\\mirae\\Documents\\Research\\Data\\Growth Curve Data\\2.13.23_flip_lacI_ind_unind_MP010_WT.csv"
data_df = pd.read_csv(data_location, sep = '\t')
for key in data_df.columns:
    data_df[key] = data_df[key].apply(lambda x: np.nan if x=="____" else float(x))
```

```python
output_file("original.html")

p = figure(title = "Growth Rate", width=1200, height=1200, y_axis_type="log",
          x_axis_label = 'Time (Min)',
           y_axis_label = 'Optical Density log(OD 600)',)
colors = ["olivedrab", "mediumturquoise", "maroon", "palevioletred"]
y_vals = ["WT_I", "WT_U", "flip_I", "flip_U"]
names = ["WT Induced", "WT Uninduced", "Capsule Induced", "Capsule Uninduced"]
for ind, val in enumerate(y_vals):
    time = [x for i,x in enumerate(data_df["Time(minutes)"]) if ~np.isnan(data_df[val][i])]
    values = [x for x in data_df[val] if ~np.isnan(x)]
    p.line(time, values, color = colors[ind], line_width = 5)
    p.circle(time, values, legend_label = names[ind], color = colors[ind], )

p.legend.location = "top_left"
p.legend.label_text_font_size = '35pt'
p.title.text_font_size = '35pt'
p.title.align = "center"
p.xaxis.axis_label_text_font_size = "30pt"
p.yaxis.axis_label_text_font_size = "30pt"

p.xaxis.major_label_text_font_size = "27pt"
p.yaxis.major_label_text_font_size = "27pt"
p.axis.axis_label_text_font_style = 'normal'
show(p)
```
